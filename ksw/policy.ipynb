### Yoon

```python
# 이미지  출력
yoon = cv2.imread('./Policy/yoon.jpg')
yoon = cv2.cvtColor(yoon, cv2.COLOR_RGB2BGR)
plt.xticks([])
plt.yticks([])
plt.imshow(yoon)
plt.show()
```

```python
yoon1 = pytesseract.image_to_string(yoon, lang = 'kor')
yoon1
# 개행삭제
yoon1_text = yoon1.replace("\n", "")
yoon1_text
```

```python
# 흑백 이미지 변환
yoon2 = cv2.cvtColor(yoon, cv2.COLOR_BGR2GRAY)
# 이진 이미지 변환(이미지, 기준값, 기준이상일때 값, 적용방법)
_, yoon2 = cv2.threshold(yoon2, 150, 255, cv2.THRESH_BINARY)
plt.xticks([])
plt.yticks([])
plt.imshow(yoon2, cmap = 'gray')
plt.show()
```

```python
yoon3 = pytesseract.image_to_string(yoon2, lang = 'kor')
yoon3
# 개행삭제
yoon3_text = moon3.replace("\n", "")
yoon3_text
```
