### 쓰잘데기 없는 문자 숫자 삭제

```python
import re
from nltk.corpus import stopwords
```

```python
def text_cleaning(text):
    # 날짜 정보 삭제
    text = re.sub(r'\(\d{4}\.\d{1,2}\.\d{1,2}\)', '', text)
    
    # 특수 문자 및 숫자 삭제
    text = re.sub(r'[^\w\s]', '', text)
    
    # 대표적인 책제, 대책 등의 제목 삭제
    text = re.sub(r'\d+ㆍ\d+대책', '', text)
    text = re.sub(r'\d+ 대팩', '', text)
    text = re.sub(r'\d+ㆍ\d+대책', '', text)
    text = re.sub(r'\d+:\d+대책', '', text)
    text = re.sub(r'은 대펴ㅣ', '', text)
    text = re.sub(r'사\d+대백', '', text)
    
    # 불용어 제거
    stop_words = set(stopwords.words("english"))
    words = text.split()
    words = [word for word in words if word not in stop_words]
    
    # 다시 문장으로 결합
    cleaned_text = " ".join(words)
```
