```python

import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
data=pd.read_csv("apart data.CSV", encoding= "EUC-KR")
```

```python
data.head(5)
```

```python
# 토큰화 방식 : 공백, 글자, 형태소 분석 으로 토큰화 한다.
from sklearn.feature_extraction.text import CountVectorizer

countVec1=CountVectorizer()
# 토큰화 후 수치화(텍스트를 숫자형태로 바꿔줌)
countVec1.fit(data['단지명'])
countVec1.vocabulary_
```

```python
from konlpy.tag import Okt

okt=Okt()
```

```python
okt.morphs(data['단지명'][2])
```

```python
data['단지명']
```

```python
tokenized_data = []
data['단지명'] = data['단지명'].astype(str)
for sentence in data['단지명']:
    tokens = okt.morphs(sentence)
    tokenized_data.append(' '.join(tokens))

data['단지명'] = tokenized_data

print(data['단지명'])
```

```python
countVec1.fit(data['단지명'])
```

```python
countVec1.vocabulary_
```
