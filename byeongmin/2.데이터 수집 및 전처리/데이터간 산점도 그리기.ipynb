```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("apart data.CSV", encoding= "EUC-KR")
```

```python
data
```

```python
years = data['계약년'].unique()
num_years = len(years)
colors = plt.cm.jet([i / float(num_years - 1) for i in range(num_years)])  # 색상 설정

plt.figure(figsize=(10, 6))  # 그래프 크기 조정

for i, year in enumerate(years):
    year_data = data[data['계약년'] == year]
    plt.scatter(year_data['층'], year_data['거래금액(만원)'], color=colors[i], label=str(year))

plt.xlabel('floor')
plt.ylabel('price(10000)')
plt.legend(title='year')
plt.grid(True)
plt.show()
```
