```python
import pandas as pd
```

```python
data= pd.read_csv("apart data.csv", encoding ="EUC-KR")
```

```python
data.head()
```

```python
data=data.drop(['번지','본번','부번','도로명','해제사유발생일','거래유형','중개사소재지'], axis =1)
```

```python
data.head()
```

```python
data['거래건수'] = data.groupby(['행정명', '계약년'])['행정명'].transform('count')

# 결과 확인
data
```

```python
data.to_csv('거래건수 추가.csv', encoding='EUC-KR')
```
