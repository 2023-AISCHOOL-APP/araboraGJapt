```python
model_lgb = lgb.LGBMRegressor(max_depth=10, min_child_weight=1, n_estimators=200, n_jobs=-1,num_leaves=30)
```

```python
def print_rmse_r2_score(model):
    # cv별 프린팅, 평균 저장
    model_name, rmse_list, r2_list = rmse_r2_cv(model)
    for i, (rmse, r2) in enumerate(zip(rmse_list, r2_list), start=1):
         print(f'{i} FOLDS: {model_name} RMSE: {rmse:.4f}, R2: {r2:.4f}')
    print(f'\n{model_name} mean RMSE: {np.mean(rmse_list):.4f}, mean R2: {np.mean(r2_list):.4f}')
    print('='*40)
    return model_name, np.mean(rmse_list), np.mean(r2_list)

def rmse_r2_cv(model):
    # cv별로 학습하는 함수
    tscv = TimeSeriesSplit(n_splits=10)
    rmse_list = []
    r2_list = []
    model_name = model.__class__.__name__
    for _, (train_index, test_index) in tqdm(enumerate(tscv.split(X), start=1), desc=f'{model_name} Cross Validations...', total=10):
        X_train, X_test = X.iloc[train_index], X.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        clf = model.fit(X_train, y_train)
        pred = clf.predict(X_test)
        rmse = RMSE(y_test, pred)
        r2 = r2_score(y_test, pred)
        rmse_list.append(rmse)
        r2_list.append(r2)
    return model_name, rmse_list, r2_list

def RMSE(y, y_pred):
    rmse = mean_squared_error(y, y_pred) ** 0.5
    return rmse
```

```python
models2 = []
rmse_scores2 = []
r2_scores2 = []
model_name, mean_rmse_score, mean_r2_score = print_rmse_r2_score(model_lgb)
models2.append(model_name)
rmse_scores2.append(mean_rmse_score)
r2_scores2.append(mean_r2_score)
```

# 시계열모델(ARIMA)

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
data['계약날짜'] = pd.to_datetime(data['계약날짜'])

# 날짜를 인덱스로 설정합니다.
data.set_index('계약날짜', inplace=True)

data.head()
```

```python
grouped = data.groupby('법정동코드')

# 각 법정동별로 ARIMA 모델을 적용하고 예측을 수행합니다.
for group, group_data in grouped:
    model = ARIMA(group_data['거래금액(만원)'], order=(5, 1, 0))
    model_fit = model.fit()
    forecast_steps = 30
    forecast = model_fit.forecast(steps=forecast_steps)
    forecast_steps2 = 60
    forecast2 = model_fit.forecast(steps=forecast_steps2)

    # 예측 결과를 출력하거나 시각화합니다.
    print(f"법정동코드: {group}")
    print("다음 30일뒤의 거래금액 예측:")
    print(forecast[:])
    print("30~ 60일뒤의 거래금액 예측:")
    print(forecast2[30:])
    
```

```python
forecast_steps_1_month = 30
forecast_1_month = model_fit.forecast(steps=forecast_steps_1_month)

# 세 달(90일) 뒤의 거래금액을 예측합니다.
forecast_steps_3_months = 90
forecast_3_months = model_fit.forecast(steps=forecast_steps_3_months)

# 예측 결과를 출력합니다.
print("한 달 뒤의 거래금액 예측:")
print(forecast_1_month[:])
print("세 달 뒤의 거래금액 예측:")
print(forecast_3_months[30:])
```
