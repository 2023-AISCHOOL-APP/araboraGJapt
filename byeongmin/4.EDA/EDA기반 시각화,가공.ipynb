```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
```

```python
font_path = r'C:\Windows\Fonts\gulim.ttf'
plt.rc('font', family='Gulim')
```

```python
data=pd.read_csv('../2~3.데이터 수집 및 전처리/머신러닝 전처리 데이터.CSV',encoding="EUC-KR")
```

```python
data=data.drop(['Unnamed: 0','계약일'],axis=1)
```

```python
data
```

```python
plt.figure(figsize=(10, 6))  # 그래프의 크기 조절 (선택사항)
plt.scatter(data['계약날짜'], data['거래금액(만원)'], color='blue', alpha=0.7)  # 산점도 그리기
plt.xlabel('계약날짜')  # x축 레이블
plt.ylabel('거래금액(만원)')  # y축 레이블
plt.title('계약날짜별 거래금액')  # 그래프 제목
plt.xticks([])  # x축 레이블 회전 (선택사항)
plt.tight_layout()  # 그래프 간격 조절 (선택사항)
plt.show()
```
