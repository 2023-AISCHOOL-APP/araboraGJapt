```python
tokenized_data = []
data['단지명'] = data['단지명'].astype(str)
for sentence in data['단지명']:
    tokens = okt.morphs(sentence)
    tokenized_data.append(' '.join(tokens))

data['단지명'] = tokenized_data

print(data['단지명'])
```
